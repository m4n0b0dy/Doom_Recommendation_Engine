```python
data_repo = '../data/json_lyrics/'
json_files = [data_repo+f for f in listdir(data_repo) if isfile(join(data_repo, f))]
```

```python
df = pd.DataFrame()
t = time.time()
for json_file in json_files:
    df = df.append(pd.DataFrame(verse_extract(json_file)), ignore_index = True)
df = df.dropna()
print('Finished extraction of all artists in',time.time()-t)
```
