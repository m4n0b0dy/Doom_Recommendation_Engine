```python
index_body = {"mappings": {
  "properties": {
    "parent_artist": {
      "type": "text"
    },
    "artist": {
      "type": "text"
    },
    "album": {
      "type": "text"
    },
    "song": {
      "type": "text"
    },
    "raw_verse": {
      "type": "text"
    },

    "verse_vector": {
      "type": "dense_vector",
      "dims": 2
    }
  }
}
             }

#delete current index
es.indices.delete(index='testing_index',
                  ignore=[400, 404])
#make index
es.indices.create(index='testing_index',
                  ignore=400,
                 body=index_body)
```

```python
es.search(index="testing_index", body={"query": {"match": {'parent_artist':'Megan Thee Stallion'}}})
```
